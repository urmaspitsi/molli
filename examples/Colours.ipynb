```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

# Colour maps:
# https://matplotlib.org/stable/tutorials/colors/colormaps.html

# Named colours:
# https://matplotlib.org/stable/gallery/color/named_colors.html

# matplotlib.axes.Axes.grid tutorials gallery:
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.grid.html
```

#### Choose colormap

```python
color_map = mpl.colormaps["twilight"]
#color_map = mpl.colormaps["Blues"]
color_map
```

```python
color_map.N
```

#### View sample of selected colours

```python
color_idxs = list(range(0, 500, 25))
num_colors = len(color_idxs)
ncols = 5
nrows = int(num_colors / ncols) if num_colors % ncols == 0 else int(num_colors / ncols) + 1
fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(10,10))
for color_idx, ax in zip(color_idxs, axs.flatten()):
    chosen_color = np.full((10, 10, 3), fill_value=color_map.colors[color_idx])
    ax.set_title(f"color_idx={color_idx}")
    ax.imshow(chosen_color)

plt.tight_layout()
plt.show()
```
